#1. Procedimiento


-**Factoriales**

El factorial es un elemento super importante para muchas técnicas de conteo, consiste en una sucesión de multiplicaciones dónde se multiplica de forma ascendente desde 1 hasta el número n, es decir, n! = 1*2*3*4*...*n. Esto se usa, por ejemplo, para la permutación, la k-permutación, o la combinación, como se verá a continuación.

Factorial usando iteraciones:
Como dijimos, el factorial consiste en multiplicar números de 1 hasta n, por tanto, en python, podemos crear esta función para definir el factorial iterando, de modo que iteramos la multiplicación desde 1 hasta n.

```python
#Definimos la función para el cálculo de un número factorial usando iteraciones
def factoriali(n):
  #Primero vemos si no salen errores, en caso de que sí, el número ingresado no
  #es un entero, por tanto, se le pide al usuario que ingrese un entero positivo
  try:
    n = int(n)

    # Verificamos que el número sea mayor o igual a 0
    if n < 0:
      print("Por favor, ingrese un número entero positivo")
      return
    # Inicializamos el resultado en 1
    resultado: int = 1

    # Iteramos desde 1 hasta n (incluyendo a n)
    for i in range(1, n + 1):
      # Multiplicamos el resultado acumulado por el número actual
      resultado = resultado * i

    # Devolvemos el cálculo final
    return resultado
  except:
    print("Por favor, ingrese un número entero positivo")
```

Factorial usando recursividad:
Sabemos que el factorial es una multiplicación sucesiva, así que también lo podemos calcular usando una función recursiva. Para esto simplemente definimos el valor de un caso base, siendo este el factorial de 0 y el factorial de 1, para luego multiplicar el número ingresado en la función por el factorial de n-1, de modo que repetirá esta operación en bucle hasta llegar a 1, dónde empezará a calcular el factorial y a multiplicarlo por el número hasta llegar a n

```python
#Definimos la función para calcular el factorial usando recursividad
def factorialr(n):
  #Al igual que en la función anterior, primero verificamos si el dato ingresado
  #Es un entero o no
  try:
    n = int(n)
  except:
    print("Por favor, ingrese un número entero positivo")
    return
  #Luego, verificamos que el número ingresado no sea menor a 0, es decir, que no
  #Sea un número negativo
  if n < 0:
    print("Por favor, ingrese un número entero positivo")
    return
  #Aquí partimos de un caso base para que la recursividad funcione
  #Sabemos que 0! y 1! siempre van a dar como resultado 1, por tanto:
  if n==0 or n==1:
    return 1
  else:
    #En este punto recurrimos a la recursividad, esto va a multiplicar n por el
    #Factorial de n-1, es decir, esta misma función. Sin embargo como esta función
    #Se va a repetir, va a llegar en algún punto hasta 1 y realizará todo el
    #cálculo hasta llegar al valor real del factorial
    return n*factorialr(n-1)
```

**-Permutacion**

La permutación es una técnica de conteo que nos permite contar la cantidad de formas de ordenar r objetos en un conjunto de n objetos. Para ello usamos la función del factorial para hallar el factorial del número n ingresado a la función, el factorial de n-r, y dividimos ambos valores.

```python
#Ahora definimos la función de permutación, que nos permitirá contar la cantidad
#De formas de ordenar r objetos distintos entre n objetos
def permutacion (n,r):
  #Para evitar cualquier error, al igual que en el factorial (Tanto el recursivo
  #como el iterativo), primero verificamos que los números sean enteros, que
  #sean mayores a 0, y adicional que r no sea mayor a n
  try:
    n = int(n)
    r = int(r)
  except:
    print("Por favor, ingrese un número entero positivo para ambos casos")
    return
  if n < 0 or r < 0:
    print("Por favor, ingrese un número entero positivo para ambos casos")
    return

  if r > n:
    print("La cantidad de objetos seleccionados no puede ser mayor a los objetos totales")
    return

  #Aquí definí n-r, debido a que estaba presentando error al hacer factoriali(n-r)
  nMenosr = n-r
  n = factoriali(n)
  #Aquí saco el factorial de n-r
  divisor = factoriali(nMenosr)
  #Y por último, la función devuelve el valor de n dividido entre n-r
  return n/divisor
```

**-Combinación**

Combinación: Esta, al igual que la permutación, es una técnica de conteo que nos permite contar la cantidad de formas en las que podemos ordenar r objetos en un conjunto de n objetos, con la diferencia de que aquí no importa el orden en que queden ordenados los objetos.

```python
#Aquí definimos la combinatoria usando el factorial y los conceptos definidos
#Anteriormente, repetimos la misma lógica que la permutación, más aquí hallamos
#el factorial de r, para poder multiplicarlo por n-r.
def combinacion(n,r):
  try:
    n = int(n)
    r = int(r)
  except:
    print("Por favor, ingrese un número entero positivo para ambos casos")
    return
  if n < 0 or r < 0:
    print("Por favor, ingrese un número entero positivo para ambos casos")
    return

  if r > n:
    print("La cantidad de objetos seleccionados no puede ser mayor a los objetos totales")
    return

  nMenosr = int(n-r)
  r = factoriali(r)
  n = factoriali(n)
  divisor = factoriali(nMenosr)
  return n/(divisor*r)
```

Comprobación de la propiedad

En las combinaciones, existe la propiedad de que para cualquier n y r enteros positivos, se cumple que la combinación(n,r) es lo mismo que la combinación(n, n,r). Esta parte del programa verifica eso para cualquier combinación.

```python
#Aquí comprobamos que se cumpla la propiedad
#nCr = nC(n-r)
def comprobar(n,r):
  #De nuevo, en todo este bloque evitamos incongruencias con respecto a
  #n y r
  try:
    n = int(n)
    r = int(r)
  except:
    print("Por favor, ingrese un número entero positivo para ambos casos")
    return
  if n < 0 or r < 0:
    print("Por favor, ingrese un número entero positivo para ambos casos")
    return
  if r > n:
    print("La cantidad de objetos seleccionados no puede ser mayor a los objetos totales")
    return

  #Hallamos ambas combinaciones usando la función de combinación anterior
  nr = combinacion(n,r)
  n_nr = combinacion(n, n-r)
  #Luego, igualamos ambos numeros calculaos anteriormente para poder verificar si la
  #propiedad se está cumpliendo.
  if nr == n_nr:
    return "Se cumple la propiedad: nCr = nC(n-r)"
  else:
    return "No se cumple que la propiedad: nCr = nC(n-r)"
```

**-Triangulo de pascal**

El triángulo de pascal se trata de un arreglo triángular de números que inicia con un 1 en la cima, y los siguientes números son una suma de los dos números encima de él. Esto se relaciona con la combinación, ya que podemos hallar el valor de los elementos en el triángulo de pascal a través de la relación combinación(n,r), dónde n es la fila del triángulo (Contando desde 0), y r es la "columna" de esa fila (Contando desde 0 también).

Fila de pascal usando combinacion: Como dijimos antes, usando la combinación(n,r), dónde contando desde 0, n es la fila y r es la "Columna" de esa fila, podemos hallar los valores que están en el triángulo de pascal, que es exactamente lo que hace este código.

```python
#Definimos la función para obtener una fila del triangulo de pascal (Contando la
#primera fila como fila 0 en esta función en específico)
def filaPascal(n):
  #de nuevo, verificamos que haya coherencia en los datos
  try:
    n = int(n)
  except:
    print("Por favor, ingrese un número entero positivo")
    return
  if n < 0:
    print("Por favor, ingrese un número entero positivo")
    return

  #Aquí definimos un array vacío, que nos permitirá formar la fila que queremos
  array = []
  #Aquí iteramos de 0 hasta n, usando nuestra función de combinación para armar
  #la fila correspondiente
  for i in range(n+1):
    array.append([int(combinacion(n,i))])
  return array
```

N primeras filas del triangulo de pascal: Aquí de nuevo aplicamos esta propiedad de la combinación que mencionamos reiteradas veces. Es importante tener en cuenta que aquí generamos las filas del triángulo de pascal contando desde 1, no desde 0 como en el cálculo de las filas. Simplemente usamos la función definida antes y la iteramos hasta n para generar las filas del triángulo deseadas.

```python
#Aquí definimos la función para formar el triangulo de pascal

def filasDePascal(n):
  #De nuevo, repetimos el mismo procedimiento de antes para evitar datos ilógicos
  try:
    n = int(n)
  except:
    print("Por favor, ingrese un número entero positivo")
    return
  if n < 0:
    print("Por favor, ingrese un número entero positivo")
    return

  #Y simplemente iteramos desde la fila 1 hasta la fila n (Tomando en cuenta
  #Que la fila 1 es la fila 0, por eso en el range vamos solo hasta n, para
  #mostrar exactamente la cantidad de filas que el usuario quiere ver)
  for i in range(n):
    print(filaPascal(i))

```
